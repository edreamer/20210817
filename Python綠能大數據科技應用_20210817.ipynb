```python
# 基礎爬蟲
# 發出要求取得原始資料
import requests
url = "http://ehappy.tw/bsdemo1.htm"
html = requests.get(url)
html.encoding = "utf-8"
print(html.text)
```

```python
# 解析網頁
from bs4 import BeautifulSoup
sp = BeautifulSoup(html.text, 'html.parser')
# sp.body.div.a.text
sp.a.text
```

```python
# 取得屬性
sp.a.get('href')
sp.a['href']
sp.img.get('src')
sp.img['src']
```
